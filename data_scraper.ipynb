```python
import numpy as np
import pandas as pd

import yfinance as yf
import investpy as inv
```

```python
sp500 = yf.download('^GSPC', interval='1mo')
dji = yf.download('^DJI', interval='1mo')
```

```python
market_index = pd.DataFrame()
market_index['SP500'] = sp500['Adj Close']
market_index['DJI'] =dji['Adj Close']
market_index.dropna(inplace=True)
market_index
```

```python
market_index.to_csv('data/market_indices.csv')
```

```python
unemployment_rate = unemployment_rate['UNRATE']
```

```python
mortgage_rate = mortgage_rate.resample('MS').mean()
```

```python
gdp
```

```python
gdp_ = pd.DataFrame(index=pd.date_range('2000-01-01','2022-06-01', freq='MS'))
gdp_['gdp'] = gdp
```

```python
gdp = gdp_.fillna(method='ffill')['gdp']
```

```python
house_under_construction
```

```python
housing_dataset = pd.DataFrame()
housing_dataset['housing_price_index'] = housing_price_index['2000-01-01':'2022-06-01']
housing_dataset['mortgage_rate'] = mortgage_rate['2000-01-01':'2022-06-01']
housing_dataset['unemployment_rate'] = unemployment_rate['2000-01-01':'2022-06-01']
housing_dataset['total_construction_spending'] = total_construction_spending['2000-01-01':'2022-06-01']
housing_dataset['population'] = population['2000-01-01':'2022-06-01']
housing_dataset['new_home_sold'] = new_home_sold['2000-01-01':'2022-06-01']
housing_dataset['iron_ore_price'] = iron_ore_price['2000-01-01':'2022-06-01']
housing_dataset['concrete_purchase_price_index'] = concrete_purchase_price_index['2000-01-01':'2022-06-01'].astype('float')
housing_dataset['building_permits'] = building_permits['2000-01-01':'2022-06-01']
housing_dataset['dji'] = dji['Adj Close']['2000-01-01':'2022-06-01']
housing_dataset['gdp'] = gdp
housing_dataset['house_under_construction'] = house_under_construction['2000-01-01':'2022-06-01']
```
